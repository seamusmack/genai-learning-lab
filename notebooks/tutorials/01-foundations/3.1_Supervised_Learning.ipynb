## __3.1.1 Introduction to Supervised Learning__

Supervised learning is a type of machine learning where the algorithm learns from labeled data, consisting of input-output pairs. In supervised learning, you train the algorithm on a dataset that includes input features and corresponding correct output labels. The goal is for the algorithm to learn the mapping or relationship between the input features and the output labels. This enables it to make predictions or decisions on new, unseen data.

## __3.1.2 Supervised Learning Categories__

There are two categories of supervised learning based on the type of target variable:

- Regression

- Classification

![link text](https://labcontent.simplicdn.net/data-content/content-assets/Data_and_AI/ML/updated/Lesson_03/Image_1.png)

## __3.1.3 Classification__

When the target variable is categorical, the prediction task is classification. For instance, predicting whether an email is spam based on its content, sender information, and subject line. Think of it like sorting mail into different boxes labeled 'spam' and 'not spam.'

## __3.1.4 Regression__

When the target variable is continuous, the prediction task is regression. For example, estimating the price of a house given its location, size, number of bedrooms, and age. Imagine you are trying to predict the price of a house based on various features, similar to how a real estate agent would appraise a property.

__Let's understand the difference between regression and classification with the use case below:__

![link text](https://labcontent.simplicdn.net/data-content/content-assets/Data_and_AI/ML/updated/Lesson_03/Image_2.png)

**Regression:**

*  **Purpose:** Predict a continuous value.

*  **Example:** Predicting the exact temperature for tomorrow.

*  **Visualization:** A thermometer scale ranging from -50°F to 170°F with an indicator pointing to 84°F.

*  **Question Addressed:** "What will be the temperature tomorrow?"

*  **Output:** A specific temperature value (e.g., 84°F).


**Classification**:

*   **Purpose:** Predict a categorical outcome.

*   **Example:** Determining whether tomorrow will be hot or cold.

*   **Visualization:** The same thermometer scale divided into two colored regions: blue for "cold" and red for "hot," with a threshold separating them.

*   **Question Addressed:** "Will it be hot or cold tomorrow?"

*   **Output**: A category (e.g., "cold" or "hot").

## __3.1.5 Applications of Supervised Learning__

Supervised learning can optimize and automate processes across industries. It enhances decision-making, reduces manual effort, and enables more accurate predictions.

1. **HR Operations**: Imagine you are a recruiter tasked with finding the best candidates for job vacancies. Supervised learning can help by shortlisting resumes based on accepted or rejected candidate profiles. This improves its ability to recommend similar profiles in the future, much like how Netflix recommends movies based on your viewing history.

2. **Finance**: In the finance industry, supervised learning can automate loan approval and improve loan risk assessment. By using historical loan data to train a model, it can predict the likelihood of a loan being repaid. This is similar to how a credit card company determines your creditworthiness before issuing a card.

3. **Emails**: Think about the spam filter in your email. It uses supervised learning to distinguish between spam and non-spam emails. By training on labeled examples of spam and non-spam emails, the model learns to predict whether new, unseen emails are spam.

4. **Manufacturing**: In manufacturing, supervised learning is used to inspect quality and classify products into different grades. For example, a factory might use a machine learning model to check for defects in products and ensure they meet quality standards, much like a quality control inspector.

5. **Maritime Industry**: In the maritime industry, forecasting combines historical events and weather conditions to develop precautionary incident management. A supervised learning technique like regression model can use past data to predict tidal currents and forecast demand and supply, reducing inventory losses. Think of it like how a weather forecast predicts rain based on past weather patterns.

6. **Fraud Protection**: Supervised learning models help detect and prevent fraud. For example, to predict fraudulent activities in financial transactions, cyber attacks on servers, and fake social media profiles, the model learns from past instances of fraud. This is akin to a detective learning from previous cases to spot new crimes.

### **1. Linear Regression**

This notebook demonstrates seven fundamental machine learning algorithms with simple implementations, clear explanations, and compelling visualizations. Each algorithm is presented with synthetic data to illustrate core concepts effectively.
m
m

- **Use case:** Predicting housing prices

- **Scenario:** A real estate company wants to estimate the market value of a house based on factors such as the number of bedrooms, square footage, location, and age of the property. Imagine you are using historical sales data to predict the price of a new house on the market.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.datasets import make_classification, make_regression
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

Linear regression finds the best-fitting straight line through data points by minimizing the sum of squared residuals. It assumes a linear relationship between input features and the target variable.

**Mathematical Foundation:** y = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ + ε

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic regression data
np.random.seed(42)
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Split the data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Create and train the model
lr_model = LinearRegression()
lr_model.fit(X_train_reg, y_train_reg)

# Make predictions
y_pred_reg = lr_model.predict(X_test_reg)

# Calculate metrics
mse = mean_squared_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

# Visualization
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_train_reg, y_train_reg, alpha=0.6, color='blue', label='Training Data')
plt.scatter(X_test_reg, y_test_reg, alpha=0.6, color='red', label='Test Data')
plt.plot(X_test_reg, y_pred_reg, color='green', linewidth=2, label='Regression Line')
plt.xlabel('Feature Value')
plt.ylabel('Target Value')
plt.title('Linear Regression: Data and Fit')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(y_test_reg, y_pred_reg, alpha=0.6)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], 'r--', linewidth=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Linear Regression Results:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.3f}")
print(f"Coefficient: {lr_model.coef_[0]:.3f}")
```

### **2. Logistic Regression**

- **Use case:** Customer purchase prediction

- **Scenario:** An e-commerce company aims to predict whether a customer will purchase a product based on their browsing history, previous purchases, and demographic data. Think of it like predicting whether a window shopper will become a buyer based on their behavior.

### Theory
Logistic regression uses the logistic function to model the probability of binary outcomes. Unlike linear regression, it outputs probabilities between 0 and 1, making it ideal for classification tasks.

**Mathematical Foundation:** P(y=1|x) = 1 / (1 + e^(-(β₀ + β₁x₁ + ... + βₙxₙ)))


```python
from sklearn.linear_model import LogisticRegression

# Generate binary classification data
np.random.seed(42)
X_log, y_log = make_classification(n_samples=500, n_features=2, n_redundant=0, 
                                   n_informative=2, n_clusters_per_class=1, random_state=42)

# Split the data
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, test_size=0.2, random_state=42)

# Scale the features
scaler_log = StandardScaler()
X_train_log_scaled = scaler_log.fit_transform(X_train_log)
X_test_log_scaled = scaler_log.transform(X_test_log)

# Create and train the model
log_model = LogisticRegression(random_state=42)
log_model.fit(X_train_log_scaled, y_train_log)

# Make predictions
y_pred_log = log_model.predict(X_test_log_scaled)
y_pred_prob = log_model.predict_proba(X_test_log_scaled)[:, 1]

# Calculate metrics
accuracy_log = accuracy_score(y_test_log, y_pred_log)

# Visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_train_log == i
    plt.scatter(X_train_log_scaled[mask, 0], X_train_log_scaled[mask, 1], 
                c=color, alpha=0.6, label=f'Class {i}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
# Decision boundary
h = 0.02
x_min, x_max = X_train_log_scaled[:, 0].min() - 1, X_train_log_scaled[:, 0].max() + 1
y_min, y_max = X_train_log_scaled[:, 1].min() - 1, X_train_log_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = log_model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
plt.colorbar(label='Predicted Probability')
plt.scatter(X_test_log_scaled[:, 0], X_test_log_scaled[:, 1], c=y_test_log, 
            cmap='RdYlBu', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary and Test Data')

plt.subplot(1, 3, 3)
plt.hist(y_pred_prob[y_test_log == 0], bins=20, alpha=0.7, label='Class 0', color='red')
plt.hist(y_pred_prob[y_test_log == 1], bins=20, alpha=0.7, label='Class 1', color='blue')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.title('Probability Distribution by Class')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Logistic Regression Results:")
print(f"Accuracy: {accuracy_log:.3f}")
print("\nClassification Report:")
print(classification_report(y_test_log, y_pred_log))
```

### **3. Naïve Bayes**

- **Use case:** Email spam detection

- **Scenario:** An email service provider wants to filter out spam emails. They use Naïve Bayes to classify emails as spam or not spam based on the words and phrases used in the emails. It's like having a smart assistant who knows which emails are junk and which ones are important.

### Theory
Naive Bayes applies Bayes' theorem with the "naive" assumption that features are conditionally independent. Despite this strong assumption, it often performs surprisingly well in practice.

**Mathematical Foundation:** P(class|features) = P(features|class) × P(class) / P(features)


```python
from sklearn.naive_bayes import GaussianNB

# Use same classification data as logistic regression
X_train_nb, X_test_nb = X_train_log, X_test_log
y_train_nb, y_test_nb = y_train_log, y_test_log

# Create and train the model
nb_model = GaussianNB()
nb_model.fit(X_train_nb, y_train_nb)

# Make predictions
y_pred_nb = nb_model.predict(X_test_nb)
y_pred_prob_nb = nb_model.predict_proba(X_test_nb)

# Calculate metrics
accuracy_nb = accuracy_score(y_test_nb, y_pred_nb)

# Visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
for i, color in enumerate(['red', 'blue']):
    mask = y_train_nb == i
    plt.scatter(X_train_nb[mask, 0], X_train_nb[mask, 1], 
                c=color, alpha=0.6, label=f'Class {i}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
# Decision boundary for Naive Bayes
h = 0.02
x_min, x_max = X_train_nb[:, 0].min() - 1, X_train_nb[:, 0].max() + 1
y_min, y_max = X_train_nb[:, 1].min() - 1, X_train_nb[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = nb_model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
plt.colorbar(label='Predicted Probability')
plt.scatter(X_test_nb[:, 0], X_test_nb[:, 1], c=y_test_nb, 
            cmap='RdYlBu', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Naive Bayes Decision Boundary')

plt.subplot(1, 3, 3)
conf_matrix = confusion_matrix(y_test_nb, y_pred_nb)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

plt.tight_layout()
plt.show()

print(f"Naive Bayes Results:")
print(f"Accuracy: {accuracy_nb:.3f}")
print("\nClassification Report:")
print(classification_report(y_test_nb, y_pred_nb))
```

### **4. K-Nearest Neighbors (KNN)**

- **Use case:** Healthcare

- **Scenario:**
KNN algorithm can be applied in healthcare sector for the early prediction of heart diseases to increase survival rate. By comparing a new patient's health indicators, like blood pressure and cholesterol, with historical data, kNN identifies patterns and predicts risks efficiently.

### Theory
KNN is a lazy learning algorithm that classifies data points based on the majority vote of their k nearest neighbors. It makes no assumptions about the underlying data distribution.

**Key Concept:** Classification is based on similarity - "Tell me who your neighbors are, and I'll tell you who you are."


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Generate data with more distinct clusters
np.random.seed(42)
X_knn, y_knn = make_classification(n_samples=400, n_features=2, n_redundant=0, 
                                   n_informative=2, n_clusters_per_class=2, random_state=42)

# Split the data
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X_knn, y_knn, test_size=0.2, random_state=42)

# Scale the features
scaler_knn = StandardScaler()
X_train_knn_scaled = scaler_knn.fit_transform(X_train_knn)
X_test_knn_scaled = scaler_knn.transform(X_test_knn)

# Test different values of k
k_values = [1, 3, 5, 7, 9, 15]
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_knn_scaled, y_train_knn)
    y_pred = knn.predict(X_test_knn_scaled)
    accuracies.append(accuracy_score(y_test_knn, y_pred))

# Use optimal k
best_k = k_values[np.argmax(accuracies)]
knn_model = KNeighborsClassifier(n_neighbors=best_k)
knn_model.fit(X_train_knn_scaled, y_train_knn)
y_pred_knn = knn_model.predict(X_test_knn_scaled)

# Calculate metrics
accuracy_knn = accuracy_score(y_test_knn, y_pred_knn)

# Visualization
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_train_knn == i
    plt.scatter(X_train_knn_scaled[mask, 0], X_train_knn_scaled[mask, 1], 
                c=color, alpha=0.6, label=f'Class {i}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(k_values, accuracies, 'bo-', linewidth=2, markersize=8)
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.title('KNN Performance vs k Value')
plt.grid(True, alpha=0.3)
plt.axvline(x=best_k, color='red', linestyle='--', label=f'Best k={best_k}')
plt.legend()

plt.subplot(2, 2, 3)
# Decision boundary
h = 0.02
x_min, x_max = X_train_knn_scaled[:, 0].min() - 1, X_train_knn_scaled[:, 0].max() + 1
y_min, y_max = X_train_knn_scaled[:, 1].min() - 1, X_train_knn_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = knn_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
plt.scatter(X_test_knn_scaled[:, 0], X_test_knn_scaled[:, 1], c=y_test_knn, 
            cmap='RdYlBu', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'KNN Decision Boundary (k={best_k})')

plt.subplot(2, 2, 4)
conf_matrix = confusion_matrix(y_test_knn, y_pred_knn)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

plt.tight_layout()
plt.show()

print(f"KNN Results:")
print(f"Best k value: {best_k}")
print(f"Accuracy: {accuracy_knn:.3f}")
print("\nClassification Report:")
print(classification_report(y_test_knn, y_pred_knn))
```

### **5. Decision Trees**

- **Use case:** Rule based Decision making

- **Scenario:** In urban planning, a Decision Tree analyzes data on population density, transport usage, and community feedback to optimize public transportation across neighborhoods. Decision Trees provide a more structured, rule-based approach to enhance transport services, improving efficiency and accessibility.

### Theory
Decision trees create a model that predicts target values by learning simple decision rules inferred from data features. The tree structure represents decisions and their possible consequences.

**Key Concept:** Recursive binary splitting based on feature values that best separate the classes.
m

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Generate data suitable for decision trees
np.random.seed(42)
X_tree, y_tree = make_classification(n_samples=400, n_features=2, n_redundant=0, 
                                     n_informative=2, n_clusters_per_class=1, 
                                     class_sep=1.5, random_state=42)

# Split the data
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(X_tree, y_tree, test_size=0.2, random_state=42)

# Create and train the model with limited depth for interpretability
tree_model = DecisionTreeClassifier(max_depth=4, random_state=42, min_samples_split=20)
tree_model.fit(X_train_tree, y_train_tree)

# Make predictions
y_pred_tree = tree_model.predict(X_test_tree)

# Calculate metrics
accuracy_tree = accuracy_score(y_test_tree, y_pred_tree)

# Visualization
plt.figure(figsize=(15, 12))

plt.subplot(2, 2, 1)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_train_tree == i
    plt.scatter(X_train_tree[mask, 0], X_train_tree[mask, 1], 
                c=color, alpha=0.6, label=f'Class {i}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
# Decision boundary
h = 0.02
x_min, x_max = X_train_tree[:, 0].min() - 1, X_train_tree[:, 0].max() + 1
y_min, y_max = X_train_tree[:, 1].min() - 1, X_train_tree[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = tree_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
plt.scatter(X_test_tree[:, 0], X_test_tree[:, 1], c=y_test_tree, 
            cmap='RdYlBu', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Tree Boundaries')

plt.subplot(2, 1, 2)
plot_tree(tree_model, feature_names=['Feature 1', 'Feature 2'], 
          class_names=['Class 0', 'Class 1'], filled=True, rounded=True, fontsize=10)
plt.title('Decision Tree Structure')

plt.tight_layout()
plt.show()

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': ['Feature 1', 'Feature 2'],
    'Importance': tree_model.feature_importances_
})

plt.figure(figsize=(8, 4))
sns.barplot(data=feature_importance, x='Feature', y='Importance')
plt.title('Feature Importance in Decision Tree')
plt.ylabel('Importance Score')
plt.show()

print(f"Decision Tree Results:")
print(f"Accuracy: {accuracy_tree:.3f}")
print(f"Tree Depth: {tree_model.get_depth()}")
print(f"Number of Leaves: {tree_model.get_n_leaves()}")
print("\nFeature Importances:")
for i, importance in enumerate(tree_model.feature_importances_):
    print(f"Feature {i+1}: {importance:.3f}")
```

###**6. Random Forests**

- **Use case:** Credit scoring

- **Scenario:** A bank evaluates the creditworthiness of loan applicants by using Random Forests to analyze various factors like credit history, income, employment status, and other financial metrics. Think of it as a group of decision trees working together to give a more accurate prediction.

### Theory
Random Forest combines multiple decision trees using bootstrap aggregating (bagging) and random feature selection. It reduces overfitting and improves generalization compared to single decision trees.

**Key Concept:** "Wisdom of crowds" - multiple weak learners create a strong learner through voting.

```python
from sklearn.ensemble import RandomForestClassifier

# Use same data as decision tree for comparison
X_train_rf, X_test_rf = X_train_tree, X_test_tree
y_train_rf, y_test_rf = y_train_tree, y_test_tree

# Create and train the model
rf_model = RandomForestClassifier(n_estimators=100, max_depth=4, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Make predictions
y_pred_rf = rf_model.predict(X_test_rf)
y_pred_prob_rf = rf_model.predict_proba(X_test_rf)

# Calculate metrics
accuracy_rf = accuracy_score(y_test_rf, y_pred_rf)

# Visualization
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
colors = ['red', 'blue']
for i, color in enumerate(colors):
    mask = y_train_rf == i
    plt.scatter(X_train_rf[mask, 0], X_train_rf[mask, 1], 
                c=color, alpha=0.6, label=f'Class {i}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
# Decision boundary
h = 0.02
x_min, x_max = X_train_rf[:, 0].min() - 1, X_train_rf[:, 0].max() + 1
y_min, y_max = X_train_rf[:, 1].min() - 1, X_train_rf[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = rf_model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
plt.colorbar(label='Predicted Probability')
plt.scatter(X_test_rf[:, 0], X_test_rf[:, 1], c=y_test_rf, 
            cmap='RdYlBu', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random Forest Decision Boundary')

plt.subplot(2, 2, 3)
feature_importance = pd.DataFrame({
    'Feature': ['Feature 1', 'Feature 2'],
    'Importance': rf_model.feature_importances_
})
sns.barplot(data=feature_importance, x='Feature', y='Importance')
plt.title('Feature Importance in Random Forest')
plt.ylabel('Importance Score')

plt.subplot(2, 2, 4)
# Compare single tree vs random forest
comparison_data = pd.DataFrame({
    'Model': ['Decision Tree', 'Random Forest'],
    'Accuracy': [accuracy_tree, accuracy_rf]
})
sns.barplot(data=comparison_data, x='Model', y='Accuracy')
plt.title('Model Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()

print(f"Random Forest Results:")
print(f"Accuracy: {accuracy_rf:.3f}")
print(f"Number of Trees: {rf_model.n_estimators}")
print("\nFeature Importances:")
for i, importance in enumerate(rf_model.feature_importances_):
    print(f"Feature {i+1}: {importance:.3f}")
print(f"\nImprovement over Decision Tree: {accuracy_rf - accuracy_tree:.3f}")
```

###**7. Support Vector Machines (SVM)**

- **Use case:** Social media monitoring

- **Scenario:**
A marketing firm uses SVM to analyze binary customer sentiments on social media platforms, focusing on classifying comments as either positive or negative. This technique effectively gauges public perception, enabling precise targeting and brand management.








### Theory
SVM finds the optimal hyperplane that separates classes with the maximum margin. It can handle non-linear relationships using kernel functions and is effective in high-dimensional spaces.

**Key Concept:** Find the decision boundary that maximizes the margin between classes while minimizing classification errors.


```python
from sklearn.svm import SVC

# Generate data with some overlap for SVM
np.random.seed(42)
X_svm, y_svm = make_classification(n_samples=300, n_features=2, n_redundant=0, 
                                   n_informative=2, n_clusters_per_class=1,
                                   class_sep=0.8, random_state=42)

# Split the data
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(X_svm, y_svm, test_size=0.2, random_state=42)

# Scale the features (important for SVM)
scaler_svm = StandardScaler()
X_train_svm_scaled = scaler_svm.fit_transform(X_train_svm)
X_test_svm_scaled = scaler_svm.transform(X_test_svm)

# Create and train models with different kernels
kernels = ['linear', 'rbf', 'poly']
svm_models = {}
accuracies_svm = {}

for kernel in kernels:
    svm = SVC(kernel=kernel, random_state=42, probability=True)
    svm.fit(X_train_svm_scaled, y_train_svm)
    y_pred = svm.predict(X_test_svm_scaled)
    svm_models[kernel] = svm
    accuracies_svm[kernel] = accuracy_score(y_test_svm, y_pred)

# Use the best performing kernel
best_kernel = max(accuracies_svm, key=accuracies_svm.get)
best_svm = svm_models[best_kernel]
y_pred_svm = best_svm.predict(X_test_svm_scaled)

# Visualization
plt.figure(figsize=(15, 10))

# Plot decision boundaries for different kernels
for i, kernel in enumerate(kernels):
    plt.subplot(2, 2, i+1)
    
    # Decision boundary
    h = 0.02
    x_min, x_max = X_train_svm_scaled[:, 0].min() - 1, X_train_svm_scaled[:, 0].max() + 1
    y_min, y_max = X_train_svm_scaled[:, 1].min() - 1, X_train_svm_scaled[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = svm_models[kernel].predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, levels=50, alpha=0.8, cmap='RdYlBu')
    plt.scatter(X_train_svm_scaled[:, 0], X_train_svm_scaled[:, 1], c=y_train_svm, 
                cmap='RdYlBu', edgecolors='black', alpha=0.7)
    
    # Highlight support vectors
    plt.scatter(X_train_svm_scaled[svm_models[kernel].support_, 0], 
                X_train_svm_scaled[svm_models[kernel].support_, 1], 
                s=100, facecolors='none', edgecolors='yellow', linewidths=2)
    
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'SVM with {kernel.upper()} kernel (Acc: {accuracies_svm[kernel]:.3f})')

plt.subplot(2, 2, 4)
kernel_comparison = pd.DataFrame({
    'Kernel': list(accuracies_svm.keys()),
    'Accuracy': list(accuracies_svm.values())
})
sns.barplot(data=kernel_comparison, x='Kernel', y='Accuracy')
plt.title('SVM Kernel Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()

print(f"SVM Results:")
print(f"Best kernel: {best_kernel}")
print(f"Best accuracy: {accuracies_svm[best_kernel]:.3f}")
print(f"Number of support vectors: {len(best_svm.support_)}")
print(f"Support vector ratio: {len(best_svm.support_) / len(X_train_svm_scaled):.3f}")

print("\nKernel Comparison:")
for kernel, accuracy in accuracies_svm.items():
    print(f"{kernel.upper()}: {accuracy:.3f}")
```
